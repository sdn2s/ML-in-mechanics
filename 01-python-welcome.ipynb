### Объекты

```python
class Student: # класс
    def __init__(self, name, age, grade): # конструктор
        self.name = name 
        self.age = age
        self.grade = grade  # средний балл

    def is_honors(self): # метод
        return self.grade >= 4.5

    def birthday(self):
        self.age += 1

    def __repr__(self): # перегрузка вывода в консоль
        return f"{self.name}, {self.age} лет, средний балл: {self.grade:.2f}"
```

```python
s = Student("Алиса", 21, 4.7)
print(s)              # Алиса, 21 лет, средний балл: 4.70
print(s.is_honors())  # True
s.birthday()
print(s.age)          # 22
```

**Упражнение (со звездочкой):** Создайте класс для арифметики комплексных чисел (перегрузите сложение, умножение, разность, деление, операции сравнения).

```python
class Complex:
    def __init__(self, re: float, im: float = 0.0):
        self.re = float(re)
        self.im = float(im)

    @staticmethod
    def _coerce(other):
        if isinstance(other, Complex):
            return other
        return Complex(other, 0.0)

    def __add__(self, other):
        other = self._coerce(other)
        return Complex(self.re + other.re, self.im + other.im)

    __radd__ = __add__

    def __sub__(self, other):
        other = self._coerce(other)
        return Complex(self.re - other.re, self.im - other.im)

    def __rsub__(self, other):
        other = self._coerce(other)
        return Complex(other.re - self.re, other.im - self.im)

    def __mul__(self, other):
        other = self._coerce(other)
        re = self.re * other.re - self.im * other.im
        im = self.re * other.im + self.im * other.re
        return Complex(re, im)

    __rmul__ = __mul__

    def __truediv__(self, other):
        other = self._coerce(other)
        denom = other.re ** 2 + other.im ** 2
        if denom == 0:
            raise ZeroDivisionError("division by zero")
        re = (self.re * other.re + self.im * other.im) / denom
        im = (self.im * other.re - self.re * other.im) / denom
        return Complex(re, im)

    def __rtruediv__(self, other):
        other = self._coerce(other)
        return other.__truediv__(self)

    def __eq__(self, other):
        other = self._coerce(other)
        return abs(self.re - other.re) < 1e-9 and abs(self.im - other.im) < 1e-9

    def __repr__(self):
        sign = "+" if self.im >= 0 else "-"
        return f"{self.re}{sign}{abs(self.im)}i"

z1 = Complex(1, 2)
z2 = Complex(3, -1)
(z1 + z2, z1 - z2, z1 * z2, z1 / z2, z1 == z1)
```

**Упражнение 15:** Нарисуйте график функции y = sin(x) и y = cos(x) на одном графике для x от 0 до 2π. Сделайте сетку, подписи осей, легенду, и залейте область между кривыми полупрозрачным цветом.

```python
x = np.linspace(0, 2*np.pi, 400)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.figure(figsize=(6,4))
plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")
plt.fill_between(x, y_sin, y_cos, color="lightblue", alpha=0.3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin(x) и cos(x)")
plt.grid(True)
plt.legend()
plt.show()
```
